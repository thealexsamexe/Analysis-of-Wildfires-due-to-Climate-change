Importing the essential library

```python
import pandas as pd
```

Reading csv file 'cleaned emission data.csv'

```python
emission_data = pd.read_csv('cleaned emission data.csv')
emission_data
```

Taking only essential columns from dataset

```python
emission_data1 = emission_data[['Country', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009',
                               '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']]
```

Checking the first five rows

```python
emission_data1.head()
```

Creating the column 'Net Emissions' and putting the sum of all the columns in it

```python
emission_data1["Net Emissions"] = emission_data1.sum(axis=1)
```

Reading the DataFrame of first five rows

```python
emission_data1.head()
```

Reading the DataFrame

```python
emission_data1
```

Sorting the values of 'Net Emissions' in descending order

```python
sorted_data = emission_data1.sort_values(by = 'Net Emissions', ascending = False)
```

reading the first 15 rows in DataFrame

```python
sorted_data.head(15)
```

Dropping the rows which is not important

```python
sorted_data = sorted_data.drop(axis = 0, index = [227, 64, 72, 13, 4, 138, 1])
```

Reading the dataframe

```python
sorted_data
```

Reading the first 12 rows

```python
sorted_data.head(12)
```

Selecting rows and columns and Reading the first 15 rows in DataFrame

```python
a = sorted_data.iloc[0:15]
a.head(15)
```

Dropping the unimportant rows

```python
a.drop(index = [14, 136, 192, 217, 167, 172], inplace = True)
```

Adding the rows 'Brazil' 

```python
b = emission_data1[emission_data1["Country"] == "Brazil"]
a = a.append(b)
```

Sorting the 'Net Emissions' in descending order and reseting the index

```python
Econ = a.sort_values(by = "Net Emissions", ascending = False)
Econ.reset_index(inplace = True)
Econ.drop(columns = ["index"], axis = 1, inplace = True)
```

Reading the dataframe

```python
Econ
```

Exporting the csv file as 'Top Emitters data.csv'

```python
Econ.to_csv('Top Emitters data.csv', index = False)
```
