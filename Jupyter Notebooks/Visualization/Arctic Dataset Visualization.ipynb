# Importing the essential libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the dataset via Pandas library function

```python
arctic_data = pd.read_csv(r"C:\Users\Sunshine\Downloads\cleaned arctic data.csv", index_col = "year")
arctic_data
```

# Plotting the Line plot of the dataset for Extent

```python
plt.figure(figsize=(14,6))
plt.title("Arctic Data")
sns.lineplot(data=arctic_data, x = arctic_data.index, y = arctic_data['extent'], marker = 'o')                                    #line chart
plt.xlabel("Year", fontsize = 20)
plt.ylabel("Extent (in Million sq.km)", fontsize = 20)
```

# Plotting the Line plot of the dataset for Area

```python
plt.figure(figsize=(14,6))
plt.title("Arctic Data")
sns.lineplot(data=arctic_data, x = arctic_data.index, y = arctic_data['area'], marker = 'o')                                    #line chart
plt.xlabel("Year", fontsize = 20)
plt.ylabel("Area (in Million sq.km)", fontsize = 20)
```

# Plotting the Line plot of the dataset 

```python
plt.figure(figsize=(14,6))
plt.title("Arctic Data")
sns.lineplot(data=arctic_data.index, x = arctic_data['area'], y = arctic_data['extent'], marker = 'o')       #line chart
plt.xlabel("Area (in Million sq.km)", fontsize = 20)
plt.ylabel("Extent (in Million sq.km)", fontsize = 20)
```
