# Bar plot for Country vs Net Emissions

```python
plt.figure(figsize=(14,6))
plt.title("Emission Data")
sns.barplot(x=emission_data1["Country"], y=emission_data1["Net Emissions"])               #bar chart
```

# Line plot for Net Emissions over the years

```python
plt.figure(figsize=(14,6))
plt.title("Emission Data")
sns.lineplot(data = emission_data1)
plt.ylabel("Net Emissions")
```

# Dropping the selected indexes

```python
a.drop(index=[14,136,192,217,167,172],inplace=True)
```

# Making of Econ Dataset

```python
b=emission_data1[emission_data1["Country"]=="Brazil"]
a=a.append(b)
Econ=a.sort_values(by="Net Emissions",ascending=False)
Econ.reset_index(inplace=True)
Econ.drop(columns=["index"],axis=1,inplace=True)
```

# Reading the Econ dataset

```python
Econ
```

# Barplot for Country vs Net emissions based on Econ dataset

```python
plt.figure(figsize=(14,6))
plt.title("Emission Data")
sns.barplot(x=Econ["Country"], y=Econ["Net Emissions"], data = Econ)
```

# Pie Chart for Country vs Net emissions based on Econ dataset

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.axis('equal')
ax.pie(Econ["Net Emissions"], labels = Econ["Country"],autopct='%1.2f%%')
plt.show()
```

# Pie Chart for Country vs Net emissions based on Econ dataset (2017)

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.axis('equal')
ax.pie(Econ["2017"], labels = Econ["Country"],autopct='%1.2f%%')
plt.show()
```
