# Importing the essential libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the dataset via Pandas library function

```python
nasa_data = pd.read_csv(r"C:\Users\Sunshine\Downloads\cleaned nasa data.csv", index_col = "year")
nasa_data
```

# Plotting the Line Plot of the dataset

```python
plt.figure(figsize=(14,6))
plt.title("NASA GLOBAL LAND-OCEAN TEMPERATURE INDEX")
sns.lineplot(data=nasa_data, marker = 'o')                                    #line chart
plt.xlabel("Year", fontsize = 20)
plt.ylabel("Temperature", fontsize = 20)
plt.savefig('image.png')
```

# Plotting the Bar Plot of Dataset for Annual Mean

```python
plt.figure(figsize=(14,6))
plt.title("NASA GLOBAL LAND-OCEAN TEMPERATURE INDEX")
sns.barplot(x=nasa_data.index, y=nasa_data["Annual_mean"])                # bar chart for annual mean
```

# Plotting the Bar Plot of the dataset for Lowess

```python
plt.figure(figsize=(14,6))
plt.title("NASA GLOBAL LAND-OCEAN TEMPERATURE INDEX")
sns.barplot(x=nasa_data.index, y=nasa_data["Lowess"])                        #bar chart for lowess
```

# Plotting the Scatterplot of the dataset

```python
plt.figure(figsize=(14,6))
plt.title("NASA GLOBAL LAND-OCEAN TEMPERATURE INDEX")
sns.scatterplot(x = nasa_data["Annual_mean"], y = nasa_data["Lowess"])             #scatterplot
```

# Plotting the Scatterplot of the dataset having a regression line

```python
plt.figure(figsize=(14,6))
plt.title("NASA GLOBAL LAND-OCEAN TEMPERATURE INDEX")
sns.regplot(x = nasa_data["Annual_mean"], y = nasa_data["Lowess"])                 #scatter plot with regression line
```

# Plotting the Histogram plot of the dataset for Annual Mean

```python
plt.figure(figsize=(14,6))
plt.title("NASA GLOBAL LAND-OCEAN TEMPERATURE INDEX")
sns.distplot(a = nasa_data["Annual_mean"], kde=False)
```

# Plotting the Histogram plot of the dataset for Lowess

```python
plt.figure(figsize=(14,6))
plt.title("NASA GLOBAL LAND-OCEAN TEMPERATURE INDEX")
sns.distplot(a = nasa_data["Lowess"], kde=False)
```

# Plotting the Histogram plot for the dataset

```python
plt.figure(figsize=(14,6))
plt.title("NASA GLOBAL LAND-OCEAN TEMPERATURE INDEX")
sns.distplot(a=nasa_data["Annual_mean"], kde=False)
sns.distplot(a=nasa_data["Lowess"], kde=False)
#plt.legend()                                     #No handlers with labels found to put in legend(o/p)
```

# Plotting the Density plot for the dataset

```python
plt.figure(figsize=(14,6))
plt.title("NASA GLOBAL LAND-OCEAN TEMPERATURE INDEX")
sns.kdeplot(data = nasa_data["Annual_mean"], shade = True)
sns.kdeplot(data = nasa_data["Lowess"], shade = True)
```
